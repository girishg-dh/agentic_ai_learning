```python
import os
import json
import pandas as pd

# Set the path to your data
DATA_DIR = "data"

# Get a list of all JSON files in the directory
json_files = [f for f in os.listdir(DATA_DIR) if f.endswith('.json')]

print(f"Found {len(json_files)} JSON files.")

# Let's inspect the first file to understand its structure
if json_files:
    first_file_path = os.path.join(DATA_DIR, json_files[0])

    print(f"--- Inspecting: {json_files[0]} ---")

    with open(first_file_path, 'r') as f:
        data = json.load(f)

    # 'json_normalize' is great for viewing nested JSON data as a flat table
    # We look inside the 'metrics' list from your JSON structure
    if 'data' in data and 'metrics' in data['data']:
        df = pd.json_normalize(data['data']['metrics'])
        display(df)
    else:
        print("Could not find the 'data.metrics' structure in this file.")
```
